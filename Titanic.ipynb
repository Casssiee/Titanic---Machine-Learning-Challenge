## Getting ready for analysis

```python
train_clean.dtypes.sort_values()
```

```python
#Remove insignificant variables
train_data = train_clean.drop(['PassengerId', 'Name', 'Embarked','Ticket', 'Fare', 'Cabin'], axis=1)
```

```python
X = train_data.iloc[:, 1:]
y = train_data['Survived']
```

# Building ensemble model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, RepeatedKFold
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
np.random.seed(0)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

np.random.seed(0)
```

```python
#specify the parameter combinations to check
param_grid = {
    'n_estimators': np.arange(1, 5, 1),
    'max_features': np.arange(1, X_train.shape[1]+1, 1)
}

#using repeated k-fold cross validation
folds = RepeatedKFold(random_state=0)

#creating the grid search object
grid_search = GridSearchCV(RandomForestClassifier(), param_grid=param_grid, cv=folds, n_jobs=-1)

#perform the grid search
grid_search.fit(X_train, y_train)
```

```python
best_rf = grid_search.best_estimator_
```

```python
y_pred = best_rf.predict(X_test)

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_clean["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train_clean[features])
X_test = pd.get_dummies(test_clean[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_clean.PassengerId, 'Survived': predictions})
output.to_csv('submission1.csv', index=False)
print("Your submission was successfully saved!")
```
